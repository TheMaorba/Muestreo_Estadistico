# Estimaci√≥n de par√°metros bajo Muestreo Estratificado
Este notebook reproduce paso a paso el ejercicio de estimaci√≥n para la regi√≥n Caribe sobre aguacates infectados por enfermedades, usando muestreo estratificado con 3 estratos.

```python
# üìå Librer√≠as
import numpy as np
```

##  Datos de las muestras
Los datos corresponden al n√∫mero de plantas infectadas por hect√°rea en 21 parcelas para cada estrato.

```python
# Datos de cada estrato (extra√≠dos del ejercicio)
estrato1 = np.array([48,53,64,62,45,47,59,65,54,45,48,46,50,63,55,57,46,64,61,54,33])
estrato2 = np.array([20,31,45,36,17,26,30,15,30,18,40,25,24,29,19,42,27,33,51,48,36])
estrato3 = np.array([74,68,77,70,72,73,78,76,69,80,74,78,71,79,77,71,79,68,75,68,77])
```

##  Tama√±os de estratos y de muestra
Definimos el tama√±o de la poblaci√≥n por estrato y el tama√±o de la muestra para cada uno.

```python
N1, N2, N3 = 270, 180, 330
N = N1 + N2 + N3
n1 = n2 = n3 = 21  # Tama√±o de muestra igual para cada estrato
print(f"Total poblaci√≥n: N={N}")
```

##  Calcular medias y varianzas muestrales por estrato
Usamos las f√≥rmulas:
- Media muestral: ùë¶ÃÑ‚Çï=‚àëùë¶/ùëõ‚Çï
- Varianza muestral: S¬≤‚Çï=Œ£(ùë¶-ùë¶ÃÑ‚Çï)¬≤/(ùëõ‚Çï-1)

```python
# Medias muestrales
mean1, mean2, mean3 = estrato1.mean(), estrato2.mean(), estrato3.mean()
# Varianzas muestrales
var1, var2, var3 = estrato1.var(ddof=1), estrato2.var(ddof=1), estrato3.var(ddof=1)
print(f"Media estrato 1: {mean1:.2f}")
print(f"Media estrato 2: {mean2:.2f}")
print(f"Media estrato 3: {mean3:.2f}")
print(f"Varianza estrato 1: {var1:.2f}")
print(f"Varianza estrato 2: {var2:.2f}")
print(f"Varianza estrato 3: {var3:.2f}")
```

## Calcular pesos de los estratos
Pesos W‚Çï=N‚Çï/N para cada estrato.

```python
W1, W2, W3 = N1/N, N2/N, N3/N
print(f"Pesos: W1={W1:.3f}, W2={W2:.3f}, W3={W3:.3f}")
```

## üìå Estimador puntual de la media estratificada
Usamos la f√≥rmula:
$$ \bar{Y}_{st} = \sum W_h \bar{y}_h $$

```python
Y_st = W1*mean1 + W2*mean2 + W3*mean3
print(f"Estimaci√≥n puntual de la media estratificada: {Y_st:.2f}")
```

## üìà Calcular varianza y desviaci√≥n est√°ndar de la media estratificada
La f√≥rmula de la varianza es:
$$ V(\bar{Y}_{st}) = \sum W_h^2 \frac{S_h^2}{n_h} \left(1-\frac{n_h}{N_h}\right) $$

```python
V_Yst = (W1**2 * var1 / n1 * (1-n1/N1) +
         W2**2 * var2 / n2 * (1-n2/N2) +
         W3**2 * var3 / n3 * (1-n3/N3))
SE_Yst = np.sqrt(V_Yst)
print(f"Varianza de la media estratificada: {V_Yst:.2f}")
print(f"Desviaci√≥n est√°ndar: {SE_Yst:.2f}")
```

##  IC al 95% para la media estratificada
Usamos el valor cr√≠tico z=1.96 para el nivel de confianza del 95%.

```python
z = 1.96
IC_media = (Y_st - z*SE_Yst, Y_st + z*SE_Yst)
print(f"IC 95% para la media estratificada: ({IC_media[0]:.2f}, {IC_media[1]:.2f})")
```

## üìå Estimaci√≥n del total y su IC al 95%
Usamos:
$$ \hat{T} = N \bar{Y}_{st} $$
$$ V(\hat{T}) = N^2 V(\bar{Y}_{st}) $$

```python
T_hat = N * Y_st
V_T = N**2 * V_Yst
SE_T = np.sqrt(V_T)
IC_total = (T_hat - z*SE_T, T_hat + z*SE_T)
print(f"Estimaci√≥n puntual del total de plantas infectadas: {T_hat:.2f}")
print(f"IC 95% para el total de plantas infectadas: ({IC_total[0]:.2f}, {IC_total[1]:.2f})")
```
