```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Simulamos una población de 1000 unidades con un valor asociado (por ejemplo, ingreso)
np.random.seed(3)
N = 1000  # tamaño de la población
poblacion = pd.DataFrame({
    'ID': np.arange(1, N+1),
    'Ingreso': np.random.normal(loc=3000, scale=500, size=N)
})

```

```python
poblacion.head()
```

```python
# Parámetros del muestreo sistemático
n = 100  # tamaño de la muestra deseada
k = N // n  # intervalo de muestreo
r = np.random.randint(1, k+1)  # punto de inicio aleatorio

# Selección sistemática
indices_muestra = np.arange(r-1, N, k)  # -1 porque Python indexa desde 0
muestra = poblacion.iloc[indices_muestra]

print(f"Tamaño de la población: {N}")
print(f"Tamaño de la muestra: {n}")
print(f"Intervalo de muestreo: {k}")
print(f"Punto de inicio aleatorio: {r}")
```

```python
muestra.head()
```

```python
# Visualización
fig, ax = plt.subplots(figsize=(12, 5))
ax.plot(poblacion['ID'], poblacion['Ingreso'], label='Población completa', alpha=0.5)
ax.scatter(muestra['ID'], muestra['Ingreso'], color='red', label='Muestra sistemática')
ax.set_title('Muestreo Aleatorio Sistemático')
ax.set_xlabel('ID de la Población')
ax.set_ylabel('Ingreso ($)')
ax.legend()
plt.tight_layout()

plt.show()
```
