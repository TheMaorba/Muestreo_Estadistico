# Estimaci贸n de Proporci贸n con Muestreo Sistem谩tico y FCP (Satisfaccci贸n de estudiantes)
Este notebook resuelve paso a paso el ejercicio comentado en clase de estudiantes satisfechos en una muestra de 18, considerando factor de correcci贸n por poblaci贸n finita (FCP).

##  F贸rmulas clave
- Proporci贸n muestral:
$$ \hat{p} = \frac{a}{n} $$
- Factor de correcci贸n:
$$ fcp^2 = \frac{N-n}{N-1} $$
- Varianza ajustada:
$$ \mathrm{Var}(\hat{p}) = \frac{\hat{p}\hat{q}}{n} \times fcp^2 $$
- Error est谩ndar:
$$ SE(\hat{p}) = \sqrt{\mathrm{Var}(\hat{p})} $$
- IC al 95%:
$$ IC = \hat{p} \pm t \cdot SE(\hat{p}) $$
- Estimaci贸n total:
$$ Total = N \cdot \hat{p} $$

```python
# Par谩metros del ejercicio actualizado
import numpy as np
import scipy.stats as st

N = 10000  # poblaci贸n
n = 18     # muestra
a = 8      # n煤mero de satisfechos
p_hat = a / n
q_hat = 1 - p_hat
print(f"Proporci贸n estimada p: {p_hat:.4f}")
print(f"Proporci贸n complementaria q: {q_hat:.4f}")
```

##  Paso 1: Calcular el factor de correcci贸n por poblaci贸n finita (fcp)

```python
fcp2 = (N - n) / (N - 1)
print(f"Factor de correcci贸n fcp虏: {fcp2:.5f}")
```

##  Paso 2: Calcular varianza y error est谩ndar ajustados con fcp

```python
var_p = (p_hat * q_hat / n) * fcp2
se_p = np.sqrt(var_p)
print(f"Varianza ajustada de p: {var_p:.5f}")
print(f"Error est谩ndar ajustado SE(p): {se_p:.5f}")
```

##  Paso 3: Calcular intervalo de confianza (IC) para la proporci贸n

```python
t_critico = st.t.ppf(0.975, df=n-1)
ic_inf = p_hat - t_critico * se_p
ic_sup = p_hat + t_critico * se_p
print(f"IC(95%) para la proporci贸n: ({ic_inf:.3f}, {ic_sup:.3f})")
```

## Paso 4: Calcular el n煤mero total estimado de estudiantes satisfechos y su IC

```python
total_est = N * p_hat
ic_total_inf = N * ic_inf
ic_total_sup = N * ic_sup
print(f"Total estimado de satisfechos: {total_est:.0f}")
print(f"IC(95%) para el total: ({ic_total_inf:.0f}, {ic_total_sup:.0f})")
```
