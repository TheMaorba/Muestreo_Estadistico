# Estimaci贸n con Factor de Correcci贸n por Poblaci贸n Finita (FCP)
Este notebook resuelve paso a paso el ejercicio de estimaci贸n del porcentaje de grasa con muestreo sistem谩tico comentado y explicado en la clase, considerando el **factor de correcci贸n por poblaci贸n finita**.

La poblaci贸n es de N=1,000 envases, y la muestra es de n=10 envases.

##  F贸rmulas clave

- Promedio muestral:
$$ \bar{x} = \frac{\sum x_i}{n} $$
- Varianza muestral:
$$ S^2 = \frac{\sum (x_i - \bar{x})^2}{n-1} $$
- Factor de correcci贸n:
$$ fcp = \sqrt{\frac{N-n}{N-1}} $$
- Varianza de la media ajustada:
$$ \mathrm{Var}(\bar{x}) = \frac{S^2}{n} \times fcp^2 $$
- Error est谩ndar:
$$ SE(\bar{x}) = \sqrt{\mathrm{Var}(\bar{x})} $$
- IC al 95%:
$$ IC = \bar{x} \pm t \cdot SE(\bar{x}) $$
- Tama帽o de muestra con precisi贸n deseada E:
$$ n = \frac{N \cdot t^2 \cdot S^2}{E^2 (N-1) + t^2 S^2} $$

```python
# Datos de la muestra y par谩metros
import numpy as np
import scipy.stats as st

datos = np.array([2.5, 2.6, 2.7, 2.6, 2.8, 2.9, 3.0, 2.6, 2.7, 2.8])
n = len(datos)
N = 1000
print(f"Datos de la muestra: {datos}")
```

##  Paso 1: Calcular el promedio muestral

```python
media = np.mean(datos)
print(f"Promedio de grasa por envase: {media:.3f}")
```

## Paso 2: Calcular la varianza muestral (S虏)

```python
varianza = np.var(datos, ddof=1)
print(f"Varianza muestral S虏: {varianza:.5f}")
```

## Paso 3: Calcular el factor de correcci贸n por poblaci贸n finita (fcp)

```python
fcp = np.sqrt((N - n) / (N - 1))
print(f"Factor de correcci贸n fcp: {fcp:.5f}")
```

##  Paso 4: Calcular varianza y error est谩ndar ajustados con fcp

```python
var_media_fcp = (varianza / n) * fcp**2
se_media_fcp = np.sqrt(var_media_fcp)
print(f"Varianza de la media ajustada: {var_media_fcp:.5f}")
print(f"Error est谩ndar ajustado: {se_media_fcp:.5f}")
```

## Paso 5: Calcular intervalo de confianza (IC) para el promedio

```python
t_critico = st.t.ppf(0.975, df=n-1)
lim_inf = media - t_critico * se_media_fcp
lim_sup = media + t_critico * se_media_fcp
print(f"IC(95%) del promedio: ({lim_inf:.3f}, {lim_sup:.3f})")
```

## Paso 6: Calcular la cantidad total estimada de grasa

```python
total_grasa = N * media
print(f"Cantidad total estimada de grasa: {total_grasa:.0f}")
```

## Paso 7: Calcular IC para el total de grasa con fcp

```python
ic_total_inf = N * lim_inf
ic_total_sup = N * lim_sup
print(f"IC(95%) para el total: ({ic_total_inf:.0f}, {ic_total_sup:.0f})")
```

## Paso 8: Calcular el tama帽o de muestra necesario con precisi贸n deseada 卤0.05

Usando la f贸rmula con fcp:

```python
E = 0.05
numerador = N * t_critico**2 * varianza
denominador = E**2 * (N-1) + t_critico**2 * varianza
n_necesario = numerador / denominador
print(f"Tama帽o de muestra necesario: {np.ceil(n_necesario):.0f}")
```
