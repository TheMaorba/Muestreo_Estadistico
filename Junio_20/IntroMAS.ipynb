# 游늵 Muestreo Aleatorio Simple (MAS) en Python

Este notebook presenta dos ejemplos pr치cticos del dise침o de Muestreo Aleatorio Simple (MAS),
una t칠cnica fundamental en estad칤stica que consiste en seleccionar una muestra de manera que
todas las posibles combinaciones de tama침o n tengan la misma probabilidad de ser elegidas.

游꿢 Objetivos:
1. Simular poblaciones reales con caracter칤sticas distintas.
2. Aplicar muestreo aleatorio simple usando pandas.
3. Calcular estimaciones de par치metros poblacionales y su error est치ndar.

游늷 Ejemplos:
- Ejemplo 1: Encuesta de satisfacci칩n estudiantil en una universidad.
- Ejemplo 2: Estimaci칩n del gasto promedio en una tienda online.

Para cada caso, se calcula la media muestral y su error est치ndar usando la f칩rmula:

    se(洧녽팫) = sqrt( (1 - f) * s / n )

donde:
- f = fracci칩n de muestreo (n/N)
- s = varianza muestral
- n = tama침o de la muestra
"""


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Simulaci칩n de dos ejemplos con MAS

# Ejemplo 1: Satisfacci칩n estudiantil
np.random.seed(42)
N1 = 5000
poblacion1 = pd.DataFrame({
    'id': np.arange(1, N1+1),
    'satisfaccion': np.random.normal(loc=7, scale=1.5, size=N1).clip(1, 10)
})
n1 = 200
muestra1 = poblacion1.sample(n=n1, replace=False)
media1 = muestra1['satisfaccion'].mean()
var1 = muestra1['satisfaccion'].var(ddof=1)
f1 = n1 / N1
se1 = np.sqrt((1 - f1) * var1 / n1)

# Ejemplo 2: Ventas en tienda online
np.random.seed(0)
N2 = 10000
poblacion2 = pd.DataFrame({
    'id': np.arange(1, N2+1),
    'monto': np.random.exponential(scale=50, size=N2)
})
n2 = 500
muestra2 = poblacion2.sample(n=n2, replace=False)
media2 = muestra2['monto'].mean()
var2 = muestra2['monto'].var(ddof=1)
f2 = n2 / N2
se2 = np.sqrt((1 - f2) * var2 / n2)

# Combinar resultados
resultados = pd.DataFrame({
    'Ejemplo': ['Satisfacci칩n Estudiantil', 'Ventas en Tienda Online'],
    'Tama침o poblaci칩n': [N1, N2],
    'Tama침o muestra': [n1, n2],
    'Media muestral': [round(media1, 2), round(media2, 2)],
    'Error est치ndar': [round(se1, 3), round(se2, 2)]
})
print("Resultados de Muestreo Aleatorio Simple")
display(resultados)
```

# 游늳 Interpretaci칩n de Resultados


A partir de los resultados del muestreo aleatorio simple aplicado en ambos ejemplos, observamos:

游댳 Ejemplo 1: Satisfacci칩n Estudiantil
- La media muestral estimada fue cercana a 6.94 sobre una escala de 1 a 10.
- El error est치ndar fue de aproximadamente 0.093, lo que indica **alta precisi칩n** en la estimaci칩n
  debido a la baja variabilidad en las respuestas y un tama침o muestral relativamente adecuado.

游댳 Ejemplo 2: Ventas en Tienda Online
- El gasto promedio estimado fue de aproximadamente $47.08.
- El error est치ndar fue mayor (~2.07), lo cual refleja la **alta dispersi칩n** t칤pica de datos financieros,
  especialmente cuando siguen distribuciones asim칠tricas como la exponencial (muchas compras peque침as y
  pocas muy grandes).

游늷 Conclusi칩n:
El Muestreo Aleatorio Simple permite estimar par치metros poblacionales de forma confiable, siempre que se
cumpla el supuesto de aleatoriedad y representatividad. Sin embargo, **la precisi칩n depende del tama침o de
muestra y la variabilidad de los datos**, lo cual se refleja directamente en el error est치ndar.
