#Ejemplo Actividad de Visisitas Diarias (Teorema Límite Central)
Un analista web quiere analizar el número de visitas diarias a un sitio web para entender las tendencias de tráfico y tomar decisiones informadas sobre campañas publicitarias.

```python
# Importar librerías necesarias
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Establecer semilla para reproducibilidad
np.random.seed(123)

# Simular visitas diarias durante un año (365 días) usando distribución exponencial
dias = 365
visitas_diarias = np.random.exponential(scale=300, size=dias)

# Calcular la media de visitas diarias y su desviación estándar poblacional
media_visitas = np.mean(visitas_diarias)
desv_estandar_poblacional = np.std(visitas_diarias, ddof=0)  # desviación estándar poblacional

print(f"Media de visitas diarias: {media_visitas:.2f}")
print(f"Desviación estándar poblacional: {desv_estandar_poblacional:.2f}")

# Visualizar la distribución de visitas diarias
plt.figure(figsize=(8, 5))
sns.histplot(visitas_diarias, bins=30, kde=True, color='blue')
plt.title("Distribución de Visitas Diarias")
plt.xlabel("Número de Visitas")
plt.ylabel("Frecuencia")
plt.grid(True)
plt.show()

# -------------------------------
# Teorema del Límite Central (TLC)
# -------------------------------

# Parámetros para simulación
num_simulaciones = 10000
tamano_muestra = 30
medias_visitas = []

for _ in range(num_simulaciones):
    muestra = np.random.choice(visitas_diarias, size=tamano_muestra, replace=True)
    medias_visitas.append(np.mean(muestra))

media_muestral = np.mean(medias_visitas)
desv_muestral = np.std(medias_visitas, ddof=1)

print(f"\nMedia de medias muestrales: {media_muestral:.2f}")
print(f"Desviación estándar de medias (empírico): {desv_muestral:.2f}")

# -------------------------------
# Cálculo del Error Estándar Teórico
# -------------------------------

error_estandar_teorico = desv_estandar_poblacional / np.sqrt(tamano_muestra)
print(f"Error estándar teórico: {error_estandar_teorico:.2f}")

# Visualizar la distribución muestral de las medias
plt.figure(figsize=(8, 5))
sns.histplot(medias_visitas, bins=30, kde=True, color='green')
plt.title("Distribución Muestral de la Media de Visitas Diarias")
plt.xlabel("Media de Visitas")
plt.ylabel("Frecuencia")
plt.grid(True)
plt.show()
```
