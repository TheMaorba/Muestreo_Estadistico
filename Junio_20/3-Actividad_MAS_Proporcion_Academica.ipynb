### üéØ Objetivo:
Aplicar el Muestreo Aleatorio Simple (MAS) para estimar la proporci√≥n de estudiantes que aprobaron un curso, y analizar la distribuci√≥n muestral de esta proporci√≥n para validar el Teorema Central del L√≠mite (TCL).

## 1. üì• Cargar el dataset simulado de notas acad√©micas

```python
# Cargar aqu√≠ los datos un archivo real si lo tienes)
import pandas as pd
import numpy as np

data = pd.read_excel('datos_academicos_simulados.xlsx')



# Mostrar las primeras filas
data.head()
```

## 2. ‚úÖ Estimar la proporci√≥n real de estudiantes aprobados

```python
prop_real = data['aprobado'].mean()
print(f"Proporci√≥n real de estudiantes aprobados: {prop_real:.3f}")
```

## 3. üéØ Estimar la proporci√≥n desde una muestra aleatoria simple

```python
n = 100
muestra = data.sample(n=n, replace=False)
prop_muestra = muestra['aprobado'].mean()

print(f"Proporci√≥n muestral estimada: {prop_muestra:.3f}")
```

```python
n = 200
muestra = data.sample(n=n, replace=True)
prop_muestra = muestra['aprobado'].mean()

print(f"Proporci√≥n muestral estimada: {prop_muestra:.3f}")
```

## 4. üìä Simular distribuci√≥n muestral de proporciones

```python
simulaciones = 1000
proporciones = []

for _ in range(simulaciones):
    muestra = data.sample(n=n, replace=False)
    proporciones.append(muestra['aprobado'].mean())

import matplotlib.pyplot as plt
import seaborn as sns

sns.histplot(proporciones, bins=30, kde=True, color='blue')
plt.axvline(prop_real, color='red', linestyle='--', label='Proporci√≥n real')
plt.title("Distribuci√≥n Muestral - Proporci√≥n de Aprobados")
plt.xlabel("Proporci√≥n muestral")
plt.ylabel("Frecuencia")
plt.legend()
plt.show()
```
