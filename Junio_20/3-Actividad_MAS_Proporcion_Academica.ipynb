### ðŸŽ¯ Objetivo:
Aplicar el Muestreo Aleatorio Simple (MAS) para estimar la proporciÃ³n de estudiantes que aprobaron un curso, y analizar la distribuciÃ³n muestral de esta proporciÃ³n para validar el Teorema Central del LÃ­mite (TCL).

## 1. ðŸ“¥ Cargar el dataset simulado de notas acadÃ©micas

```python
# Cargar aquÃ­ los datos un archivo real si lo tienes)
import pandas as pd
import numpy as np

data = pd.read_excel('datos_academicos_simulados.xlsx')



# Mostrar las primeras filas
data.head()
```

## 2. âœ… Estimar la proporciÃ³n real de estudiantes aprobados

```python
prop_real = data['aprobado'].mean()
print(f"ProporciÃ³n real de estudiantes aprobados: {prop_real:.3f}")
```

## 3. ðŸŽ¯ Estimar la proporciÃ³n desde una muestra aleatoria simple

```python
n = 100
muestra = data.sample(n=n, replace=False)
prop_muestra = muestra['aprobado'].mean()

print(f"ProporciÃ³n muestral estimada: {prop_muestra:.3f}")
```

```python
n = 200
muestra = data.sample(n=n, replace=True)
prop_muestra = muestra['aprobado'].mean()

print(f"ProporciÃ³n muestral estimada: {prop_muestra:.3f}")
```

## 4. ðŸ“Š Simular distribuciÃ³n muestral de proporciones

```python
simulaciones = 1000
proporciones = []

for _ in range(simulaciones):
    muestra = data.sample(n=n, replace=False)
    proporciones.append(muestra['aprobado'].mean())

import matplotlib.pyplot as plt
import seaborn as sns

sns.histplot(proporciones, bins=30, kde=True, color='blue')
plt.axvline(prop_real, color='red', linestyle='--', label='ProporciÃ³n real')
plt.title("DistribuciÃ³n Muestral - ProporciÃ³n de Aprobados")
plt.xlabel("ProporciÃ³n muestral")
plt.ylabel("Frecuencia")
plt.legend()
plt.show()
```
