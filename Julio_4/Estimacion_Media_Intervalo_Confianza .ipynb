# üìä Estimaci√≥n de la Media e Intervalo de Confianza

A partir de los datos de minutos de programa por media hora de transmisi√≥n sin comerciales, calcular la estimaci√≥n puntual de la media y construir un intervalo de confianza del 95% usando Muestreo Aleatorio Simple (MAS).

image.png

## 1. üì• Ingresar los datos proporcionados

```python
# Datos en minutos
import numpy as np
datos = np.array([
    21.06, 22.24, 20.62, 21.66, 21.23, 23.86,
    23.82, 20.30, 21.52, 22.51, 21.91, 23.11,
    20.02, 22.20, 21.20, 22.37, 22.19, 22.34,
    23.36, 23.44
])
n = len(datos)
n
```

## 2. ‚úÖ Estimaci√≥n puntual de la media y la desviaci√≥n est√°ndar muestral

```python
media_muestral = np.mean(datos)
s_muestral = np.std(datos, ddof=1)

print(f"Media muestral estimada: {media_muestral:.2f} minutos")
print(f"Desviaci√≥n est√°ndar muestral: {s_muestral:.2f} minutos")
```

## 3. üìê F√≥rmula para el error est√°ndar y el intervalo de confianza

El **error est√°ndar de la media** es:

$\
SE(\bar{y}) = \frac{s}{\sqrt{n}}
\$

El **intervalo de confianza del 95%** para la media es:

$\
\bar{y} \pm t_{\alpha/2, n-1} \times SE(\bar{y})
\$

donde $t_{\alpha/2, n-1}$ es el valor cr√≠tico de la distribuci√≥n t de Student con $n-1$ grados de libertad.

## 4. üßÆ Calcular el intervalo de confianza del 95%

```python
from scipy.stats import t

SE = s_muestral / np.sqrt(n)
alpha = 0.05
t_critico = t.ppf(1 - alpha/2, df=n-1)

limite_inferior = media_muestral - t_critico * SE
limite_superior = media_muestral + t_critico * SE

print(f"Error est√°ndar: {SE:.3f}")
print(f"Valor cr√≠tico t: {t_critico:.3f}")
print(f"Intervalo de confianza 95%: ({limite_inferior:.2f}, {limite_superior:.2f}) minutos")
```

## 5. üß† Interpretaci√≥n

Podemos decir con un 95% de confianza que el tiempo promedio real de minutos de programa por media hora de transmisi√≥n sin comerciales se encuentra entre los l√≠mites calculados. Esto significa que, si repiti√©ramos el muestreo muchas veces, el 95% de los intervalos construidos contendr√≠an la verdadera media poblacional.
