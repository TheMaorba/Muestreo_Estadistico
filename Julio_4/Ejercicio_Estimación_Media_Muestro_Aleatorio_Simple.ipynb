```python
# Importar base de datos
import pandas as pd
df = pd.read_csv('social_media_engagement.csv')
df.head()
```

```python
# Tama√±o de la poblacion
N = df.shape[0]
print(f'Tama√±o de la poblaci√≥n: {N}')
```

```python
#Muestra aleatoria simple de tama√±o 120
n = 120
muestra = df.sample(n=n, replace=False, random_state=42)
muestra.head()
```

```python
#Internalo de confianza al 95% para el promedio de minutos diarios en la app time_in_app_min
import numpy as np
from scipy.stats import t

# Media poblacional
media_pob_min = df['time_in_app_min'].mean()
# Media y desviaci√≥n en la muestra
media_muestra_min = muestra['time_in_app_min'].mean()
s_muestra_min = muestra['time_in_app_min'].std(ddof=1)

# Calcular IC
SE_math = s_muestra_min / np.sqrt(n)
t_critico = t.ppf(0.975, df=n-1)
IC_inf_math = media_muestra_min - t_critico * SE_math
IC_sup_math = media_muestra_min + t_critico * SE_math

print(f'Media horas: {media_pob_min:.2f}')
print(f'IC 95% para los minutos en la app: ({IC_inf_math:.2f}, {IC_sup_math:.2f})')

```

```python
#Internalo de confianza al 95% para el promedio de likes avg_likes
import numpy as np
from scipy.stats import t

# Media poblacional
media_pob_likes = df['avg_likes'].mean()
# Media y
media_muestra_likes = muestra['avg_likes'].mean()
s_muestra_likes = muestra['avg_likes'].std(ddof=1)

#
SE_math = s_muestra_likes / np.sqrt(n)
t_critico = t.ppf(0.975, df=n-1)
IC_inf_math = media_muestra_likes - t_critico * SE_math
IC_sup_math = media_muestra_likes + t_critico * SE_math

print(f'Media likes: {media_pob_likes:.2f}')
print(f'IC 95% para los likes: ({IC_inf_math:.2f}, {IC_sup_math:.2f})')

```

```python
# Compara las medias de las muestras con las medias poblacionales. ¬øLas estimaciones parecen precisas?
print(f'Media poblacional de minutos en la app: {media_pob_min:.2f}')
print(f'Media muestral de minutos en la app: {media_muestra_min:.2f}')
print(f'Media poblacional de likes: {media_pob_likes:.2f}')
print(f'Media muestral de likes: {media_muestra_likes:.2f}')
```

Los datos nos muestran las medias poblacionales y muestrales para dos m√©tricas: minutos en la aplicaci√≥n y "me gusta". Para los minutos en la aplicaci√≥n, la media poblacional es de 119.79 minutos, mientras que la media muestral es de 115.93 minutos. En cuanto a los "me gusta", la media poblacional es de 302.59, y la media muestral es de 301.18. En ambos casos, se observa que la media muestral es ligeramente inferior a la media poblacional, lo cual es un comportamiento esperado cuando se utiliza una muestra para estimar las caracter√≠sticas de una poblaci√≥n m√°s grande.

```python
#Realiza 300 muestras aleatorias de tama√±o 40 para construir la distribucion muestral del promedio de likes. Calcula la media y desviacion de esa distribucion.

import numpy as np
#Realiza 300 muestras aleatorias de tama√±o 40 para construir la distribucion muestral del promedio de likes. Calcula la media y desviacion de esa distribucion.
np.random.seed(42) # Para reproducibilidad
num_muestras = 300
tamano_muestra = 40
medias_muestrales_likes = []

for _ in range(num_muestras):
  muestra_likes = df['avg_likes'].sample(n=tamano_muestra, replace=False)
  media_muestra_likes_sim = muestra_likes.mean()
  medias_muestrales_likes.append(media_muestra_likes_sim)

# Calcula la media y desviaci√≥n de la distribuci√≥n muestral
media_distribucion_muestral = np.mean(medias_muestrales_likes)
desviacion_distribucion_muestral = np.std(medias_muestrales_likes, ddof=1)

print(f'Media de la distribuci√≥n muestral del promedio de likes: {media_distribucion_muestral:.2f}')
print(f'Desviaci√≥n de la distribuci√≥n muestral del promedio de likes: {desviacion_distribucion_muestral:.2f}')
```

Aqu√≠ te explico c√≥mo esta parte del c√≥digo **evidencia** el TCL, incluso sin una visualizaci√≥n expl√≠cita:

1.  **Generaci√≥n de M√∫ltiples Muestras:** El c√≥digo toma 300 muestras aleatorias de tama√±o 40 de la variable `avg_likes`.
2.  **C√°lculo de las Medias Muestrales:** Para cada una de estas 300 muestras, calcula la media de `avg_likes`.
3.  **Distribuci√≥n de las Medias:** Al almacenar estas 300 medias muestrales en la lista `medias_muestrales_likes`, estamos creando una aproximaci√≥n de la **distribuci√≥n muestral de la media**.

**¬øC√≥mo evidencia esto el TCL?**

El TCL establece que si tomas muestras lo suficientemente grandes (generalmente se considera n >= 30) de cualquier poblaci√≥n, independientemente de su distribuci√≥n original, la distribuci√≥n de las medias muestrales se aproximar√° a una **distribuci√≥n normal**.

Aunque el c√≥digo no grafica esta distribuci√≥n, el hecho de que est√©s calculando la media y desviaci√≥n est√°ndar de `medias_muestrales_likes` (que son la media y desviaci√≥n est√°ndar de la distribuci√≥n muestral) te permite hacer inferencias basadas en el TCL. Si graficaras esta lista, ver√≠as una forma que se parecer√≠a a una campana (distribuci√≥n normal), y la media de esta distribuci√≥n muestral ser√≠a muy cercana a la media poblacional de `avg_likes`. La desviaci√≥n est√°ndar de esta distribuci√≥n muestral es una estimaci√≥n del error est√°ndar de la media.

**En resumen:** La simulaci√≥n, al generar y recopilar las medias de m√∫ltiples muestras, est√° creando los datos que, seg√∫n el TCL, deben seguir una distribuci√≥n normal. Aunque la normalidad no se *muestra* directamente, el proceso de calcular la media y desviaci√≥n de `medias_muestrales_likes` se basa en la expectativa que el TCL nos da sobre c√≥mo se comportan estas medias.
