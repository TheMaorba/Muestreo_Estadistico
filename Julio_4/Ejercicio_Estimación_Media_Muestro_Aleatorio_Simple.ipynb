```python
# Importar base de datos
import pandas as pd
df = pd.read_csv('social_media_engagement.csv')
df.head()
```

```python
# Tamaño de la poblacion
N = df.shape[0]
print(f'Tamaño de la población: {N}')
```

```python
#Muestra aleatoria simple de tamaño 120
n = 120
muestra = df.sample(n=n, replace=False, random_state=42)
muestra.head()
```

```python
#Internalo de confianza al 95% para el promedio de minutos diarios en la app time_in_app_min
import numpy as np
from scipy.stats import t

# Media poblacional
media_pob_min = df['time_in_app_min'].mean()
# Media y desviación en la muestra
media_muestra_min = muestra['time_in_app_min'].mean()
s_muestra_min = muestra['time_in_app_min'].std(ddof=1)

# Calcular IC
SE_math = s_muestra_min / np.sqrt(n)
t_critico = t.ppf(0.975, df=n-1)
IC_inf_math = media_muestra_min - t_critico * SE_math
IC_sup_math = media_muestra_min + t_critico * SE_math

print(f'Media horas: {media_pob_min:.2f}')
print(f'IC 95% para los minutos en la app: ({IC_inf_math:.2f}, {IC_sup_math:.2f})')

```

```python
#Internalo de confianza al 95% para el promedio de likes avg_likes
import numpy as np
from scipy.stats import t

# Media poblacional
media_pob_likes = df['avg_likes'].mean()
# Media y
media_muestra_likes = muestra['avg_likes'].mean()
s_muestra_likes = muestra['avg_likes'].std(ddof=1)

#
SE_math = s_muestra_likes / np.sqrt(n)
t_critico = t.ppf(0.975, df=n-1)
IC_inf_math = media_muestra_likes - t_critico * SE_math
IC_sup_math = media_muestra_likes + t_critico * SE_math

print(f'Media likes: {media_pob_likes:.2f}')
print(f'IC 95% para los likes: ({IC_inf_math:.2f}, {IC_sup_math:.2f})')

```

```python
# Compara las medias de las muestras con las medias poblacionales. ¿Las estimaciones parecen precisas?
print(f'Media poblacional de minutos en la app: {media_pob_min:.2f}')
print(f'Media muestral de minutos en la app: {media_muestra_min:.2f}')
print(f'Media poblacional de likes: {media_pob_likes:.2f}')
print(f'Media muestral de likes: {media_muestra_likes:.2f}')
```

Los datos nos muestran las medias poblacionales y muestrales para dos métricas: minutos en la aplicación y "me gusta". Para los minutos en la aplicación, la media poblacional es de 119.79 minutos, mientras que la media muestral es de 115.93 minutos. En cuanto a los "me gusta", la media poblacional es de 302.59, y la media muestral es de 301.18. En ambos casos, se observa que la media muestral es ligeramente inferior a la media poblacional, lo cual es un comportamiento esperado cuando se utiliza una muestra para estimar las características de una población más grande.

```python
#Realiza 300 muestras aleatorias de tamaño 40 para construir la distribucion muestral del promedio de likes. Calcula la media y desviacion de esa distribucion.

import numpy as np
#Realiza 300 muestras aleatorias de tamaño 40 para construir la distribucion muestral del promedio de likes. Calcula la media y desviacion de esa distribucion.
np.random.seed(42) # Para reproducibilidad
num_muestras = 300
tamano_muestra = 40
medias_muestrales_likes = []

for _ in range(num_muestras):
  muestra_likes = df['avg_likes'].sample(n=tamano_muestra, replace=False)
  media_muestra_likes_sim = muestra_likes.mean()
  medias_muestrales_likes.append(media_muestra_likes_sim)

# Calcula la media y desviación de la distribución muestral
media_distribucion_muestral = np.mean(medias_muestrales_likes)
desviacion_distribucion_muestral = np.std(medias_muestrales_likes, ddof=1)

print(f'Media de la distribución muestral del promedio de likes: {media_distribucion_muestral:.2f}')
print(f'Desviación de la distribución muestral del promedio de likes: {desviacion_distribucion_muestral:.2f}')
```

El código proporcionado no simula directamente el Teorema Central del Límite (TCL) de manera explícita mediante la visualización de la distribución de las medias muestrales. Sin embargo, el proceso de tomar múltiples muestras y calcular sus medias para luego analizar la distribución de estas medias (`medias_muestrales_likes`) **es el núcleo de lo que el TCL describe**.

Aquí te explico cómo esta parte del código **evidencia** el TCL, incluso sin una visualización explícita:

1.  **Generación de Múltiples Muestras:** El código toma 300 muestras aleatorias de tamaño 40 de la variable `avg_likes`.
2.  **Cálculo de las Medias Muestrales:** Para cada una de estas 300 muestras, calcula la media de `avg_likes`.
3.  **Distribución de las Medias:** Al almacenar estas 300 medias muestrales en la lista `medias_muestrales_likes`, estamos creando una aproximación de la **distribución muestral de la media**.

**¿Cómo evidencia esto el TCL?**

El TCL establece que si tomas muestras lo suficientemente grandes (generalmente se considera n >= 30) de cualquier población, independientemente de su distribución original, la distribución de las medias muestrales se aproximará a una **distribución normal**.

Aunque el código no grafica esta distribución, el hecho de que estés calculando la media y desviación estándar de `medias_muestrales_likes` (que son la media y desviación estándar de la distribución muestral) te permite hacer inferencias basadas en el TCL. Si graficaras esta lista, verías una forma que se parecería a una campana (distribución normal), y la media de esta distribución muestral sería muy cercana a la media poblacional de `avg_likes`. La desviación estándar de esta distribución muestral es una estimación del error estándar de la media.

**En resumen:** La simulación, al generar y recopilar las medias de múltiples muestras, está creando los datos que, según el TCL, deben seguir una distribución normal. Aunque la normalidad no se *muestra* directamente, el proceso de calcular la media y desviación de `medias_muestrales_likes` se basa en la expectativa que el TCL nos da sobre cómo se comportan estas medias.
